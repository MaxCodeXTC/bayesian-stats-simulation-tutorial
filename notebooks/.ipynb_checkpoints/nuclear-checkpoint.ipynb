```python
import numpy as np
import pandas as pd

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('nuclear_power_accidents.csv')
df.Date = pd.to_datetime(df.Date)
df = df[df.Date >= pd.to_datetime('1974-01-01')]

inter_times = np.diff(np.sort(df.Date)).astype(float) / 1e9 / 3600 / 24
```

```python
x, y = np.sort(inter_times), np.arange(1, len(inter_times)+1) / len(inter_times)

p = bokeh.plotting.figure(height=250, width=350, 
                          x_axis_label='inter-incident time (days)', 
                          y_axis_label='ECDF')
p.circle(x, y)
bokeh.io.show(p)
```
